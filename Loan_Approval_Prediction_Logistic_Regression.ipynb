```python
df = pd.get_dummies(df,columns=["Property_Area"], drop_first=True,dtype=int) # To convert Categorical data into One-Hot-Encoding.
```

```python
df.head()
```

```python
df.columns
```

```python
x = df.drop("Loan_Status",axis=1)
y = df["Loan_Status"]
```

```python
x.head()
```

```python
y.head()
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
# feature scaling to resolve(ConvergenceWarning:
#lbfgs failed to converge (status=1):STOP: TOTAL NO. OF ITERATIONS REACHED LIMIT.)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred)

print(accuracy)
```

```python
# Confusion Matrix,Precision, Recall aur F1 Score(TP, TN, FP, FN).
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)

print(cm)
```

```python
from sklearn.metrics import precision_score

precision = precision_score(y_test, y_pred)

print(precision)
```

```python
from sklearn.metrics import f1_score

f1 = f1_score(y_test, y_pred)

print(f1)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
